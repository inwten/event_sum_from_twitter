```python
def f7(seq):
    seen = set()
    seen_add = seen.add
    return [x for x in seq if not (x in seen or seen_add(x))]
```

```python
import ast
my_file = open("./page_rank.txt", "r", encoding="utf-8")
  
# reading the file
data = my_file.read()
  
# replacing end splitting the text 
# when newline ('\n') is seen.
data_into_list = data.split("\n")
# print(json.dumps(data_into_list[0] , indent = 3))
page_rank = []
for i in range(len(data_into_list)):
    # print(i)
    try:
        page_rank.append(ast.literal_eval(data_into_list[i]))
    except:
        continue
my_file.close()
```

```python
entities = []
cluster = []
weight = []
for i in range(len(df)):
    entities.append(ast.literal_eval(df.Entity[i]))
    cluster.append(ast.literal_eval(df.cluster[i]))
    weight.append(ast.literal_eval(df.weight[i]))
```

```python
import ast
my_file = open("./cluster_of_word.txt", "r", encoding="utf-8")
  
# reading the file
data = my_file.read()
  
# replacing end splitting the text 
# when newline ('\n') is seen.
data_into_list = data.split("\n")
# print(json.dumps(data_into_list[0] , indent = 3))
my_file.close()
cluster_of_word = []
for i in range(len(data_into_list)):
    # print(i)
    try:
        cluster_of_word.append(ast.literal_eval(data_into_list[i]))
    except:
        continue
```

```python
def replace(lst, old_value, new_value):
    for i in range(len(lst)):
        if lst[i] == old_value:
            lst[i] = new_value
    return lst
```

```python
def summarization(top_tweet,no_of_element):
    Who = []
    What = []
    Where = []
    When = []
    for i in range(len(top_tweet)):
        nlp_ed = nlp(top_tweet[i][0])
        for ent in nlp_ed.ents:
            if (ent.label_ == 'PERSON' or ent.label_ == 'ORG'  
                or ent.label_ == 'NORP'  or ent.label_ == 'GPE_WHO' ):
                if ent.text.lower() in cluster_of_word[cluster_no]:
                    Who.append((ent.text.lower().replace(",",""),page_rank[cluster_no][ent.text.lower()]))
                else:
                    Who.append((ent.text.lower().replace(",",""),0))
            if (ent.label_ == 'ACT'):
                if ent.text.lower() in cluster_of_word[cluster_no]:
                    What.append((ent.text.lower().replace(",",""),page_rank[cluster_no][ent.text.lower()]))
                else:
                    What.append((ent.text.lower().replace(",",""),0))
            if (ent.label_ == 'GPE' or ent.label_ == 'LOC'):
                # print(ent.text)
                if ent.text.lower() in cluster_of_word[cluster_no]:
                    Where.append((ent.text.lower().replace(",",""),page_rank[cluster_no][ent.text.lower()]))
                else:
                    Where.append((ent.text.lower().replace(",",""),0))
            if (ent.label_ == 'DATE' or ent.label_ == 'TIME'):
                if ent.text.lower() in cluster_of_word[cluster_no]:    
                    When.append((ent.text.lower().replace(",",""),page_rank[cluster_no][ent.text.lower()]))
                else:
                    When.append((ent.text.lower().replace(",",""),0))
    Who = sorted(f7(Who), key=operator.itemgetter(1))[-no_of_element:]
    Who.reverse()
    What = sorted(f7(What), key=operator.itemgetter(1))[-no_of_element:]
    What.reverse()
    Where = sorted(f7(Where), key=operator.itemgetter(1))[-no_of_element:]
    Where.reverse()
    When = sorted(f7(When), key=operator.itemgetter(1))[-no_of_element:]
    When.reverse()
    return Who,What,Where,When
```

```python
Who,What,Where,When = summarization(top_tweet,10)
```
